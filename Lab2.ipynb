```python
import cv2
import numpy as np

def read_images_from_folders(directory_path, peoples_names):
    all_images = []
    all_labels = []
    label = 0
    
    for person_name in peoples_names:
        folder_path = os.path.join(directory_path, person_name)
        print(folder_path)
        images = []
        labels = []
        for filename in os.listdir(directory_path):
            if filename.endswith('.jpg') or filename.endswith('.jpeg'):
                img_path = os.path.join(folder_path, filename)
                img = cv2.imread(img_path, cv2.IMREAD_COLOR)  # Read image in color
                if img is not None:
                    images.append(img)
                    labels.append(label)
        all_images.extend(images)
        all_labels.extend(labels)
        label += 1
    
    return np.array(all_images), np.array(all_labels)

# Path to the base folder containing subfolders of people's names
#base_folder = "Dataset/Faces"

# Read images from the folders specified in the people_names list
images, labels = read_images_from_folders(directory_path, peoples_names)

# Print the shape of images and labels arrays
print("Shape of images array:", images.shape)
print("Shape of labels array:", labels.shape)
```
